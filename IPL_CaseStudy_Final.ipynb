```python
import matplotlib.pyplot as plt
import pandas as pd
# import numpy as np
df = pd.read_csv(r"C:\Nistha\Ctag Python\Datasets\IPL.csv")
# print(df.head())
print(df.isnull())
```

```python

def one():
    f = plt.figure(figsize=(20,5))
    a = df["player_of_match"].value_counts() 
    a = a.head(5)
    print(a)
    plt.title("Most player of the match")
    # plt.xlabel("Player")
    # plt.ylabel("Awards")
    a.plot(kind="bar",xlabel="player",ylabel="Awards")
    # plt.bar(a.index,a)
    plt.show()
one()
```

```python
def two():
    f = plt.figure(figsize=(20,5))
    a = df.groupby(["winner","season"]).size().unstack()
    # print(a.to_string())
    # a.plot(kind="bar",stacked=True)
    a.plot(kind="bar")
    plt.title("Matches won by each team in each season")
    plt.ylabel("Matches won")
    # plt.show()
two()
```

```python
def seven():
    f = plt.figure(figsize=(20,5))
    plt.title("Player of the  vs venue")
    plt.xlabel("player_of_match")
    plt.ylabel("winners")
    a = df.groupby("venue")["player_of_match"].count().sort_index()
    # a = df.groupby("venue")["player_of_match"].count().sort_values()
    plt.bar(a.index,a)
    # print(a)
    plt.xticks(a.index,rotation="vertical")
    plt.show()
seven()
```
